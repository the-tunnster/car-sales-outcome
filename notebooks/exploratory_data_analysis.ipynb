# Exploratory Data Analysis.

```python
numeric_columns = ["age_of_vehicle_years",
				   "sched_serv_warr", "non_sched_serv_warr", "sched_serv_paid", "non_sched_serv_paid",
				   "total_paid_services", "total_services", "mth_since_last_serv", "annualised_mileage",
				   "num_dealers_visited", "num_serv_dealer_purchased"]

categorical_columns= ["age_band", "gender", "car_model", "car_segment"]
```

```python

for col in numeric_columns:
  print(col, "Information : ")
  print("Mean : ", dataFrame[col].mean())
  print("Median : ", dataFrame[col].median())
  print("Mode : ", dataFrame[col].mode())

  print("Minimum : ", dataFrame[col].min())
  print("Maximum : ", dataFrame[col].max())

  print("Quantiles : ", numpy.quantile(dataFrame[col], [0, 0.25, 0.5, 0.75, 1]))

  print("Variance : ", numpy.var(dataFrame[col], ddof=1))
  print("Standard Deviation : ", numpy.std(dataFrame[col], ddof=1))

  print("Skew : ", dataFrame[col].skew())
  print("Kurtosis : ", dataFrame[col].kurtosis())
  print()
```

```python
ydata_profiling.ProfileReport(dataFrame)
```

```python
def cross_table(column_name: str):	
	column_dict = {}

	for i in range (dataFrame.shape[0]):
		column_value = str(dataFrame[column_name][i])
		result = dataFrame['Target'][i]

		if column_dict.get(column_value) is None:
			column_dict[column_value] = {
				'yes' : 0,
				'no' : 0
			}

		if result == 1:
			column_dict[column_value]['yes'] = column_dict[column_value]['yes'] + 1
		else:
			column_dict[column_value]['no'] = column_dict[column_value]['no'] + 1

	return (column_dict)
```

```python
for col in categorical_columns:
	column_dict = cross_table(col)
	print(col, ": \tyes\tno")
	for value in column_dict:
		print(value, "\t", column_dict[value]['yes'], "\t", column_dict[value]['no'])
	print()
```
