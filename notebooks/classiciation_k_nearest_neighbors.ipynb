```python
import numpy
import pandas
import pickle

from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
dataFrame = pandas.read_csv("../data/processed/encoded.csv")
print(dataFrame.shape)
```

```python
Y = dataFrame.pop("Target")

df_train, df_test, y_train, y_test = train_test_split(dataFrame, Y, test_size=0.2)
```

# Baseline.

```python
y_mode = Y.mode()

y_base = numpy.full(y_train.shape, y_mode)
accuracy_score(y_train, y_base)
```

# Training.

```python
model_file_path = "../models/"
model_file_name = "knn.pickle"
```

```python
knn = pickle.load(open(model_file_path + model_file_name, "rb"))
```

```python
knn = KNeighborsClassifier(n_neighbors=3,							#int, default=5
						   weights="distance",						#{‘uniform’, ‘distance’}, None				
						   algorithm="auto", 						#{‘auto’, ‘ball_tree’, ‘kd_tree’, ‘brute’}
						   leaf_size=30,							#int, default=30
)
```

```python
knn.fit(df_train, y_train)
y_train_preds = knn.predict(df_train)


print("Training score : ", accuracy_score(y_train_preds, y_train))
print("\nConfusion Matrix for Training :\n", confusion_matrix(y_train, y_train_preds, labels=knn.classes_))
```

```python
pickle.dump(knn, open(model_file_path + model_file_name, "wb"))
```

```python
y_test_preds = knn.predict(df_test)

print("Testing score : ", accuracy_score(y_test_preds, y_test))
print("\nConfusion Matrix for Testing :\n", confusion_matrix(y_test, y_test_preds, labels=knn.classes_))
```
