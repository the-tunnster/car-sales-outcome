```python
import pandas
import numpy
```

```python
base_file_path = "../data/"
```

```python
dataFrame = pandas.read_csv(base_file_path + "raw/training.csv")
```

```python
print(dataFrame.shape)
```

```python
numeric_columns = ["age_of_vehicle_years",
				   "sched_serv_warr", "non_sched_serv_warr", "sched_serv_paid", "non_sched_serv_paid",
				   "total_paid_services", "total_services", "mth_since_last_serv", "annualised_mileage",
				   "num_dealers_visited", "num_serv_dealer_purchased"]

categorical_columns= ["age_band", "gender", "car_model", "car_segment"]
```

```python
for col in dataFrame.columns:
  if col in categorical_columns:
    continue
  print(col, ": ")

  print("\tMean : ", dataFrame[col].mean())
  print("\tMedian : ", dataFrame[col].median())

  print("\tMinimum : ", dataFrame[col].min())
  print("\tMaximum : ", dataFrame[col].max())

  print()
```

```python
dataFrame[dataFrame.isna().any(axis=1)]
```

```python
for col in dataFrame.columns:
  nan_counts = dataFrame[col].isna().sum()
  if nan_counts > 0:
    print(col, ":", nan_counts)
```

```python
genderDictionary = { numpy.nan : "Unspecified"}

ageBandDictionary = {numpy.nan : "Unspecified"}
```

```python
dataFrame["gender"].replace(genderDictionary, inplace=True)
dataFrame["age_band"].replace(ageBandDictionary, inplace=True)
```

```python
for col in dataFrame.columns:
  nan_counts = dataFrame[col].isna().sum()
  if nan_counts > 0:
    print(col, ":", nan_counts)
```

```python
dataFrame.to_csv(base_file_path + "/processed/cleaned.csv", index=False)
```
